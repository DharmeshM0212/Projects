**PRICE PREDICTION MODEL USING ANN**

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/House Price India.csv')
```

```python
x=df.drop('Price',axis=1)
```

```python
x
```

```python
y=df['Price']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.4,random_state=15)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
reg=Sequential() #follows sequential order either in forward or backward
reg.add(Dense(22,activation='relu'))
reg.add(Dense(12,activation='relu'))
reg.add(Dense(32,activation='relu'))
reg.add(Dense(1,activation='linear'))
```

```python
reg.compile(optimizer='adam', loss= 'mse', metrics=['mse'])  
```

```python
reg.fit(xtrain,ytrain,batch_size=100,epochs=20,validation_data=(xtest,ytest))
```

```python
reg.predict([[9992818905,42498,7,3,3080,6000,3,1,4,5,10,4000,800,9800,0,121003,32.8895,-124.585,4000,20080,2,30]])    
```
